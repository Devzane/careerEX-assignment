## **Question 6**

Create two Series, `A` and `B`, containing the numbers 1 - 5 and 6 - 10 respectively. Compute the sum, difference ```(B-A)```, and product of `A` and `B`.

````python

# Creating Series A (numbers 1 to 5)
A = pd.Series(np.arange(1, 6))

# Creating Series B (numbers 6 to 10)
B = pd.Series(np.arange(6, 11))

# Computing the sum of A and B
sum_AB = A + B
print("Sum of A and B:\n", sum_AB)

# Computing the difference of B and A (B - A)
diff_BA = B - A
print("\nDifference of B and A (B - A):\n", diff_BA)

# Computing the product of A and B
product_AB = A * B
print("\nProduct of A and B:\n", product_AB)
````

## **Question 7**

Lists 10 attributes/methods of a Series and explain what they do.

### 1.index:

What it does: This attribute returns the index labels of the Series. The index is a fundamental part of a Series, allowing you to access and align data. It can consist of integers, strings, dates, or any other hashable Python object.

### 2. values:

What it does: This attribute returns a NumPy array containing the data values of the Series. It essentially extracts the underlying data without the index labels.

### 3. dtype:

What it does: This attribute returns the data type of the elements in the Series. Pandas automatically infers the data type, but you can also specify it explicitly during Series creation.

### 4. shape:

What it does: This attribute returns a tuple representing the dimensionality of the Series. For a 1-dimensional Series, the tuple will contain a single element: the number of elements in the Series.

### 5. size:

What it does: This attribute returns the number of elements in the Series. It's equivalent to the first (and only) element of the shape attribute.


### 6. head(n):

What it does: This method returns the first n rows of the Series. It's useful for quickly inspecting the beginning of a large Series. The default value of n is 5.

### 7. tail(n):

What it does: This method returns the last n rows of the Series. It's useful for inspecting the end of a large Series. The default value of n is 5.

### 8. value_counts(normalize=False, sort=True, ascending=False, bins=None):

What it does: This method returns a Series containing the counts of unique values in the Series.
normalize=True returns the relative frequencies (proportions) instead of counts.
sort=True sorts the resulting Series by value count (descending by default).
ascending=True sorts the resulting Series in ascending order.
bins can be used to group values into bins for a histogram-like output.

### 9. sort_values(axis=0, ascending=True, inplace=False, kind='quicksort', na_position='last', ignore_index=False, key=None):

What it does: This method sorts the values of the Series.
ascending=True sorts in ascending order (default).
ascending=False sorts in descending order.
inplace=True modifies the Series in place (returns None).
ignore_index=True resets the index to a simple integer index after sorting.

### 10. map(arg, na_action=None):

What it does: This method is used to substitute values in a Series based on a mapping correspondence. The arg can be a dictionary, a function, or another Series.
If arg is a dictionary, keys in the dictionary are replaced with their corresponding values.
If arg is a function, the function is applied to each value in the Series.
If arg is another Series, the index of the mapping Series is used to align and substitute values.
na_action='ignore' preserves NaN values.


## **Question 8**

Lists 10 attributes/methods of a DataFrame and explain what they do

## Attributes (Provide information about the DataFrame):

1. index: This attribute returns the row index labels of the DataFrame. Similar to a Series index, it provides a way to identify and access rows.

2. columns: This attribute returns the column index labels of the DataFrame. These are the names of the different variables or features in your tabular data.

3. values: This attribute returns a NumPy array representing the data values within the DataFrame. It provides the raw data without the row and column labels.

4. dtypes: This attribute returns a Series where the index is the column names, and the values are the data type of each column (e.g., integer, float, object, datetime).

5. shape: This attribute returns a tuple representing the dimensions of the DataFrame as (number of rows, number of columns).

6. size: This attribute returns the total number of elements in the DataFrame, which is the product of the number of rows and the number of columns.   

## Methods (Perform operations on the DataFrame):

7. head(n): This method returns the first n rows of the DataFrame. It's useful for a quick look at the data, with the default being the first 5 rows.

8. tail(n): This method returns the last n rows of the DataFrame, allowing you to inspect the end of the dataset. The default is the last 5 rows.

9. info(): This method provides a concise summary of the DataFrame, including the index type, column data types, the number of non-null values in each column, and memory usage.

10. describe(): This method generates descriptive statistics of the numerical columns in the DataFrame, such as count, mean, standard deviation, minimum, maximum, and percentiles.   


## **Question 9**

Highlight 3 differences between a `Series` and `DataFrame`

1. Dimensionality: A Series is one-dimensional. It's like a single column of data with an associated index (labels for each row). A DataFrame is two-dimensional. It's like a table with rows and columns, where each column can be thought of as a Pandas Series.   

2. Structure: A Series has two main components: an index and the data itself. A DataFrame has three main components: a row index, column names, and the data organized in columns. You can think of a DataFrame as a collection of Series objects that share the same index.

3. Homogeneity of Data: While both Series and DataFrames can hold various data types, a Series ideally contains data of a single data type for efficiency. A DataFrame is more flexible and can have columns of different data types within the same structure (e.g., one column of integers, another of strings).
